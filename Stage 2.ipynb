### Cleaning / Preparing Data for Model
- Grouped by player, adding up thier all nba selections 

```python
df = pd.read_csv("nba_stats.csv")
df = df.drop(['Unnamed: 0'], axis=1)
```

```python
df.columns
```

```python
df["3P%"].fillna(df["3P%"].mean(), inplace=True)
df["PER"].fillna(df["3P%"].mean(), inplace=True)
df["TS%"].fillna(df["3P%"].mean(), inplace=True)
df["3PAr"].fillna(df["3P%"].mean(), inplace=True)
df["FTr"].fillna(df["3P%"].mean(), inplace=True)
df["ORB%"].fillna(df["3P%"].mean(), inplace=True)
df["DRB%"].fillna(df["3P%"].mean(), inplace=True)
df["TRB%"].fillna(df["3P%"].mean(), inplace=True)
df["AST%"].fillna(df["3P%"].mean(), inplace=True)
df["STL%"].fillna(df["3P%"].mean(), inplace=True)
df["BLK%"].fillna(df["3P%"].mean(), inplace=True)
df["USG%"].fillna(df["3P%"].mean(), inplace=True)
df["WS/48"].fillna(df["3P%"].mean(), inplace=True)
df["FG%"].fillna(df["3P%"].mean(), inplace=True)
df["2P%"].fillna(df["3P%"].mean(), inplace=True)
df["eFG%"].fillna(df["3P%"].mean(), inplace=True)
df["FT%"].fillna(df["3P%"].mean(), inplace=True)
df["TOV%"].fillna(df["TOV%"].mean(), inplace=True)
```

```python
grouped_by_player1 = df.groupby(["Player"])
```

```python
mean_stats = grouped_by_player1['G', 'GS', 'MP', 'PER', 'TS%',
       '3PAr', 'FTr', 'ORB%', 'DRB%', 'TRB%', 'AST%', 'STL%', 'BLK%', 'TOV%',
       'USG%', 'OWS', 'DWS', 'WS', 'WS/48', 'OBPM', 'DBPM', 'BPM', 'VORP',
       'FG', 'FGA', 'FG%', '3P', '3PA', '3P%', '2P', '2PA', '2P%', 'eFG%',
       'FT', 'FTA', 'FT%', 'ORB', 'DRB', 'TRB', 'AST', 'STL', 'BLK', 'TOV',
       'PF', 'PTS'].sum()
```

```python
all_nba_sum = grouped_by_player1['All_NBA'].sum()
```

```python
all_nba_sum.head()
```

```python
mean_stats.head()
```

```python
df1 = mean_stats.join(all_nba_sum)
```

```python
df1.head()
```

```python
df2 = df1.loc[df1["PTS"] >= 1200, :].reset_index(drop = True)
df2 = df1.loc[df1["GS"] >= 120, :].reset_index(drop = True)
df2 = df1.loc[df1["MP"] >= 2000, :].reset_index(drop = True)
df2 = df1.loc[df1["G"] >= 200, :].reset_index(drop = True)
```

```python
df2['All_NBA'].value_counts()
```

```python
df2.to_csv('nba_stats2.csv')
```

```python
updated_stats = df2[['MP','G','PTS','AST','2PA','FGA','WS','All_NBA']]
```

```python
updated_stats = updated_stats.reset_index(drop = True)
updated_stats.rename(columns = {'2PA':'TwoPA', 'FT%':'FTP'}, inplace = True) 
```

```python
updated_stats.to_csv('nba_stats3.csv')
```

```python
target = updated_stats['All_NBA'].values.reshape(-1, 1)
data = updated_stats.drop("All_NBA", axis=1)

from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(data, target, random_state=42)

# from sklearn.preprocessing import StandardScaler
# X_scaler = StandardScaler().fit(X_train)
# y_scaler = StandardScaler().fit(y_train)

# X_train_scaled = X_scaler.transform(X_train)
# X_test_scaled = X_scaler.transform(X_test)
# y_train_scaled = y_scaler.transform(y_train)
# y_test_scaled = y_scaler.transform(y_test)



from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error


model = LinearRegression()

model.fit(X_train, y_train)
training_score = model.score(X_train, y_train)
testing_score = model.score(X_test, y_test)

predictions = model.predict(X_test)
MSE = mean_squared_error(y_test, predictions)
r2 = model.score(X_test, y_test)

print(f"MSE: {MSE}, R2: {r2}")
print(f"Training Score: {training_score}")
print(f"Testing Score: {testing_score}")



# LASSO model
# Note: Use an alpha of .01 when creating the model for this activity
from sklearn.linear_model import Lasso

lasso = Lasso(alpha=.01).fit(X_train, y_train)

training_score1 = lasso.score(X_train, y_train)
testing_score1 = lasso.score(X_test, y_test)

predictions1 = lasso.predict(X_test)
MSE = mean_squared_error(y_test, predictions1)
r2 = lasso.score(X_test, y_test)

print(f"MSE: {MSE}, R2: {r2}")
print(f"Training Score: {training_score1}")
print(f"Testing Score: {testing_score1}")



# Ridge model
# Note: Use an alpha of .01 when creating the model for this activity
from sklearn.linear_model import Ridge

ridge = Ridge(alpha=.01).fit(X_train, y_train)

training_score2 = ridge.score(X_train, y_train)
testing_score2 = ridge.score(X_test, y_test)

predictions2 = ridge.predict(X_test)
MSE = mean_squared_error(y_test, predictions2)
r2 = ridge.score(X_test, y_test)

print(f"MSE: {MSE}, R2: {r2}")
print(f"Training Score: {training_score2}")
print(f"Testing Score: {testing_score2}")
```
